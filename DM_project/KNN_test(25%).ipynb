```python
import pandas as pd
```

```python
file_path = '/mnt/c/Users/KDT21/Desktop/DM/diabetes.csv'
diabetes_df = pd.read_csv(file_path)
```

```python
print(diabetes_df.describe())
```

```python
#insulin 컬럼 제거
col_to_drop = 'Insulin'
non_ins_df = diabetes_df.drop(columns=[col_to_drop], axis=1)
print(non_ins_df.describe())
```

```python
# 0 데이터 값 있는 행 제거

import numpy as np
import pandas as pd

# 원본 복사, 0을 NaN으로 변환
nan_df = non_ins_df.copy()
cols_to_replace = ['Glucose', 'BloodPressure', 'SkinThickness', 'BMI']
nan_df[cols_to_replace] = nan_df[cols_to_replace].replace(0, np.nan)

# NaN 포함 행 전체 삭제
# how='any' 해당 행에 단 하나의 NaN이라도 있으면 삭제하라는 의미
dropped_df = nan_df.dropna(how='any')

print("--- NaN 포함 행 삭제 후 데이터셋 크기 ---")
print(dropped_df.shape)

print("\n--- 삭제 후 기술 통계량 (min 값 확인) ---")
print(dropped_df.describe())
```

```python
# dropped_df에서 target, input 데이터 분리
data_target = dropped_df['Outcome']
data_input = dropped_df.drop('Outcome', axis=1)
```

```python
print(data_target.head())
print(data_input.head())
```

```python
#모델 학습 전, train, test 분리
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    data_input, data_target, stratify = data_target, random_state=42)

print(train_input.shape, test_input.shape)
```

```python
# 1번, KNeighborsClassifier(non_scaled data)
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()

kn.fit(train_input, train_target)

print("---------------KNN(n=5) : non_scaled_data--------------")
print()
print(f"train score: {kn.score(train_input, train_target)}")
print(f"test score: {kn.score(test_input, test_target)}")
```

```python
# 1번, KNeighborsClassifier(non_scaled data)
# KNN 최적의 n 찾기
best_score = 0
best_n = 0

for n in range(5, 50):
    kn.n_neighbors = n
    kn.fit(train_input, train_target) 
    score = kn.score(test_input, test_target) 
    
    if score > best_score:
        best_score = score
        best_n = n

print(f"가장 높은 test 점수: {best_score:.4f}, 최적의 n 값: {best_n}")
```

```python
# 1번, KNeighborsClassifier(non_scaled data)
# n= 38로 학습
kn_b = KNeighborsClassifier(n_neighbors = 38)
kn_b.fit(train_input, train_target)

print("---------------KNN(best n=38): non_scaled_data--------------")
print()
print(f"train score: {kn_b.score(train_input, train_target)}")
print(f"test score: {kn_b.score(test_input, test_target)}")
```

```python
# 2번, KNeighborsClassifier(scaled data)
# standardscaler로 데이터 표준화(평균0, 표준편차1)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_scaled = scaler.fit_transform(train_input)  #train_data / fit and transform, scaler에 평균과 표준편차 저장
test_scaled = scaler.transform(test_input)   #test_data / transform
```

```python
print(train_scaled.shape, test_scaled.shape)
print(train_scaled)
```

```python
# 2번, KNeighborsClassifier(scaled data)
# 표준화한 데이터로 KNN model 생성
kn_scaled = KNeighborsClassifier()
kn_scaled.fit(train_scaled, train_target)

print("---------------KNN(n=5): scaled data--------------")
print()
print(f"train score: {kn_scaled.score(train_scaled, train_target)}")
print(f"test score: {kn_scaled.score(test_scaled, test_target)}")
```

```python
# 2번, KNeighborsClassifier(scaled data)
# 최적의 n 찾기
best_score = 0
best_n = 0

for n in range(5, 50):
    kn_scaled.n_neighbors = n
    kn_scaled.fit(train_scaled, train_target) 
    score = kn_scaled.score(test_scaled, test_target) 
    
    if score > best_score:
        best_score = score
        best_n = n

print(f"가장 높은 test 점수: {best_score:.4f}, 최적의 n 값: {best_n}")
```

```python
# 2번, KNeighborsClassifier(scaled data)
# n=40로 모델 train, test
kn_scaled_b = KNeighborsClassifier(n_neighbors = 40)
kn_scaled_b.fit(train_scaled, train_target)

print("---------------KNN(best n=40): scaled_data--------------")
print()
print(f"train score: {kn_scaled_b.score(train_scaled, train_target)}")
print(f"test score: {kn_scaled_b.score(test_scaled, test_target)}")
```

```python
# 3번, 하이퍼파라미터 튜닝(GridSearchCV, non_scaled data) - 파라미터 탐색 및 교차검증을 알아서 실행해줌
from sklearn.model_selection import GridSearchCV

#탐색할 매개변수와 값 생성
params = {'n_neighbors': range(3, 50),\
         'weights': ['uniform', 'distance'], \
         'metric': ['euclidean', 'manhattan']}
```

```python
gs = GridSearchCV(KNeighborsClassifier(), params, n_jobs=-1)
gs.fit(train_input, train_target)
```

```python
best = gs.best_estimator_  #params 객체에서 탐색한 매개변수 설정들 중에서 가장 최적이라고 판단된 값 저장
print(f"최적의 매개변수: {best}") 
print(f"최적의 train score: {gs.score(train_input, train_target)}")
print(f"최적의 test score: {gs.score(test_input, test_target)}")
```

```python
# 4번, 하이퍼파라미터 튜닝(GridSearchCV, scaled data) 

gs_scaled = GridSearchCV(KNeighborsClassifier(), params, n_jobs=-1)
gs_scaled.fit(train_scaled, train_target)
bestt = gs.best_estimator_  
print(f"최적의 매개변수: {bestt}") 
print(f"최적의 train score: {gs_scaled.score(train_scaled, train_target)}")
print(f"최적의 test score: {gs_scaled.score(test_scaled, test_target)}")
```
