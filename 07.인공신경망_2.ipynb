```python
import keras
(train_input, train_target), (test_input, test_target) = \
    keras.datasets.fashion_mnist.load_data()
```

```python
from sklearn.model_selection import train_test_split

train_scaled = train_input / 255.0
train_scaled = train_scaled.reshape(-1, 28*28)
train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size = 0.2, random_state = 42
)
```

```python
inputs = keras.layers.Input(shape = (784, ))
dense1 = keras.layers.Dense(100, activation = 'sigmoid', name = 'hidden_layer')  #은닉층, 몇개에 뉴런을 둘 지는 경험에 따라 결정,, 출력층보다는 많이
dense2 = keras.layers.Dense(10, activation = 'softmax', name = 'output_layer')   #출력층
```

```python
model = keras.Sequential([inputs, dense1, dense2])  #신경망 구조 정의
```

```python
model.summary()   #input layer는 계산 및 가중치가 없는 층이기때문에 summay에 안나옴
```

```python
model = keras.Sequential([
    keras.layers.Input(shape=(784,)),
    keras.layers.Dense(100, activation = 'sigmoid', name='은닉층'), 
    keras.layers.Dense(10, activation = 'softmax', name = '출력층')
], name = '패션 MNIST 모델')
```

```python
model.summary()
```

```python
model = keras.Sequential()
model.add(keras.layers.Input(shape=(784,)))
model.add(keras.layers.Dense(100, activation='sigmoid'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(train_scaled, train_target, epochs=5)
```
